# Outlier Treatment

```python
#number of counts and data type of that variable
df.info()
```

We definitely know that rainfall and n_hot_rooms contains outliers from the graph, we will write a function to treat them.

Before we do capping and flooring, we need to identify the 99th and 1st percentile values of these two variables.

To do that, we have a function in numpy called percentile.

# Outlier Treatment for n_hot_rooms

```python
#parameters: column name and the percentile we want to see
np.percentile(df.n_hot_rooms, [99])
```

But, the output here is an array, to fetch the first number of an array,w e have to specify the location of that value in a square bracket.

```python
np.percentile(df.n_hot_rooms, [99]) [0]
#fetching the first element of the previous array                          
```

```python
uv = np.percentile(df.n_hot_rooms, [99]) [0]
```

Now, we have to identify the rows where the n_hot_rooms value is more than the above number.

```python
df[(df.n_hot_rooms>uv)]
# we are getting all the data where the n_hot_rooms value is greater than this 99
```

```python
#capping the values
df.n_hot_rooms[(df.n_hot_rooms>3*uv)] = 3*uv #limit and cap the values
```

After the exceution of the above cell, run the cell before it again and you will see the outlier gone.

#Outlier Treatment for rainfall

```python
np.percentile(df.rainfall, [1]) [0]
#since the outlier is on the lower end
```

```python
lv = np.percentile(df.rainfall, [1]) [0]
```

```python
df[(df.rainfall<lv)]
```

```python
#capping the values
#lower values should be multiplied by a decimal point
df.rainfall[(df.rainfall<0.3*lv)] = 0.3*lv #limit and cap the values
```

# Variable Transformation and Deletion

```python
#scatter plot of crime rate and price
sns.jointplot(x = "crime_rate", y = "price", data= df)
```

The curve looks like a logarithmic curve and we need to tranform this curve to have a linear relationship between X and Y.

One way to do this is:
* If we take the log of the x which is the crime_rate, we can get a kind of linear relationship of crime rate with our y variable.

Since most of the values are near 0 and log(0) is not defined and it tends towards minus infinity. So to remove this, we will add a value of 1 to our crime_rate.

```python
#transform the variable
df.crime_rate = np.log(1+df.crime_rate)
```

```python
#scatter plot of crime rate and price
sns.jointplot(x = "crime_rate", y = "price", data= df)
```

Always try to make relationship between X and Y variable, a linear relationship.

The above actions is another way to treat outliers.
