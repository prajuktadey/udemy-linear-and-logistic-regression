# Seaborn

Seaborn is a library for data visualization. Most used library is matplotlib but for this course, we will be using seaborn.

```python
import seaborn as sns
```

```python
#plot the distribution of our age variable
sns.distplot(data2.Age)

#this is first creating bins of all the ages 
#the line that goes through the boxes is also known as KDE or kernl density estimate 
```

```python
#remove the KDE
sns.distplot(data2.Age, kde = False)
```

```python
help(sns.distplot)
```

```python
sns.distplot(data2.Age, kde= False, color = "red")
```

```python
#seaborn comes with various datasets
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.shape #150 rows, 5 columns
```

```python
iris.describe()
```

```python
sns.jointplot(x= "sepal_length", y= "sepal_width", data= iris)
#top we have distribution of sepal length
#side we have the distribution of sepal width
```

```python
sns.pairplot(iris)
#scatter plot for all the variables
```
